# **[1] 행, 열 조회**

## **기본적인 방법**
- 행

    - 한 개

       - `데이터프레임명[인덱스:인덱스+1]`

   - 여러 개

        - `데이터프레임명[시작인덱스:끝인덱스+1]`

```python
df.head(3)
```

```python
df[:10]
```

```python
df[3:7]
```

```python
df[3:4]
```

- 열

   - 한 개

      - `데이터프레임[컬럼명]`
      - `데이터프레임.컬럼명`

   - 여러 개

      - `데이터프레임[[컬럼명1, 컬럼명2, ...]]`

```python
df['Survived']
```

```python
df.Survived
```

```python
df['Survived'].to_frame()
```

```python
df[['Survived']]
```

```python
df[['Survived','Pclass','Name']]
```

## **loc, iloc 사용하기**

- loc: 레이블값을 사용하여 조회

     - `데이터프레임명.loc[행조건,열조건]`

     - 열만 조회할때는 행조건에 `:`를 입력합니다.

```python
df.loc[3,]
```

```python
df.loc[:,'Name']
```

```python
df.loc[3,'Name']
```

```python
df.loc[3:5,]
```

```python
df.loc[3:5,['Age','Pclass']]
```

```python
df.loc[[1,3,5],['Age','Pclass']]
```

```python
df1 = df.set_index('Name') #Name열을 인덱스로 설정
df1.head()
```

```python
df1.loc['Heikkinen, Miss. Laina',]
```

```python
df1.loc[['Heikkinen, Miss. Laina','Allen, Mr. William Henry',]]
```

```python
df1.loc[3:5,]

# 타잎 오류
```

- iloc: 위치인덱스를 사용하여 조회

    - `데이터프레임명.iloc[행인덱스조건,열인덱스조건]`

```python
df1.iloc[3,]
```

```python
df.iloc[3:7,]
```

```python
df.iloc[[1,3,5],]
```

```python
df1.iloc[3:5,]
```

```python
df.iloc[3:7,2:4]
```

# **[2] 원하는 조건으로 데이터 추출하기**

## **데이터 정렬하기**

    - `데이터프레임명.sort_values(정렬기준컬럼)`

    - 내림차순으로 정렬하고 싶다면 `ascending=False` 조건을 걸어주면 됩니다.

```python
df.head()
```

```python
df.sort_values('Age')
# 'Age' 기준으로 정렬이 된다
```

```python
df.sort_values('Age', ascending=False)
```

```python
df.sort_values(['Age','Fare'], ascending=[False,True])
```

### **특정 조건을 충족하는 데이터 추출하기**

    - `데이터프레임명[조건식]`

    - `데이터프레임명.query('조건식')`



```python
df[df['Pclass'] == 1]
```

```python
df[(df['Pclass'] == 1) & (df['Age'] >= 30)]
```

```python
df[(df['Pclass'] == 1) | (df['Age'] >= 30)]
```

```python
df.query('Pclass == 1')
```

```python
df.query('Pclass == 1 and Age >= 30')
```

```python
df.query('Pclass == 1 or Age >= 30')
```

```python
df[df['PassengerId'].isin([3,100,500])]
```

```python
df.query('PassengerId in [3,100,500]')
```

```python
df.query('PassengerId.isin([3,100,500])')
```

```python
import random

passengerid_list = list(df['PassengerId'])
passengerid_sample = random.sample(passengerid_list, 10)
```

```python
df[df['PassengerId'].isin(passengerid_sample)]
```

```python
df.query('PassengerId in @passengerid_sample')
```
