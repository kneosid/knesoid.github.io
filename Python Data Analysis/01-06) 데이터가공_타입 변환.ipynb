## **[2] 타입 변환**

`데이터명[컬럼명].astype(타입)`

```python
df['PassengerId']
```

```python
df['PassengerId'].astype(str)
```

```python
df1 = df.copy()

df1['PassengerId'] = df1['PassengerId'].astype(str)
df1.info()
```

```python
df['Age']
```

```python
# df['Age'].astype(int)
# 결측치를 변환할수 없다고 오류발생
```

```python
df['Age'].fillna(-1).astype(int)
```

```python
df1 = df.copy()

df1['Age'] = df1['Age'].fillna(-1).astype(int)
df1.tail()
```

```python
import numpy as np

df1 = df.copy()
df1['Age'] = df1['Age'].fillna(-1).astype(int).replace(-1, np.nan)
df1.tail()
```
