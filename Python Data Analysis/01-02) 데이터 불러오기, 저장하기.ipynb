## **[2] 엑셀 파일**
- 불러오기
    - `데이터변수 = pd.read_excel(파일경로, sheet_name=시트이름)`
    - 특정 시트만 선택해서 읽어올수 있다
    
- 저장하기
    - `데이터변수.to_excel(파일경로, sheet_name=시트이름)`

```python
excel_file_path = './data/titanic_train.xlsx'
excel_data = pd.read_excel(excel_file_path, sheet_name='시트1')
```

```python
excel_data.head(3)
```

- `header`: 컬럼 이름으로 사용할 행
- `index_col`: 인덱스로 사용할 컬럼
- `usecols`: 사용할 컬럼

```python
excel_data = pd.read_excel(excel_file_path, 
                           sheet_name='시트1', 
                           header=1)
# 1번 행이 컬럼으로 설정

excel_data.head(3)
```

```python
excel_data1 = pd.read_excel(excel_file_path, 
                            sheet_name='시트1', 
                            header=1, 
                            index_col = 'PassengerId', 
                            usecols = ['PassengerId',
                                       'Survived',
                                       'Pclass',
                                       'Age'])
excel_data1.head()
```

```python
excel_file_path1 = './data/excel_test.xlsx'
excel_data1.to_excel(excel_file_path1, 
                     sheet_name = 'sheet1')
# 편집한 파일을 새로운 엑셀 파일로 저장
```

```python
pd.read_excel(excel_file_path1, 
              sheet_name='sheet1').head(3)
```
