## **stack, unstack**

 - stack : 컬럼 레벨에서 인덱스 레벨로 데이터프레임을 변경합니다.

 - unstack : 인덱스 레벨에서 컬럼 레벨로 데이터프레임을 변경합니다.

      - stack의 반대입니다.

```python
import pandas as pd

file_path = './data/titanic_train.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
pivot = pd.pivot_table(df, index=['Sex','Pclass'], 
                       values=['Survived','Fare'], 
                       aggfunc=['mean','median','sum'])
pivot
```

```python
pivot.stack(0, future_stack=True)   
# 첫번째 컬럼이 인덱스로 내려갔다
```

```python
pivot.stack(1, future_stack=True)
```

```python
pivot.unstack(0) #인덱스의 첫번째 레벨을 컬럼으로 쌓아 올립니다
```

```python
pivot.unstack(1) #인덱스의 두번째 레벨을 컬럼으로 쌓아 올립니다
```

## **melt**

- `pd.melt(데이터명, id_vars=기준 컬럼)`

![](2024-06-19-18-42-15.png)


```python
data = pd.DataFrame({'name':['a','b','c']
                    , 'order_count':[3,4,10]
                    , 'amount':[10000,25000,300000]})
```

```python
data
```

```python
pd.melt(data, id_vars=['name'])
```

```python
pd.melt(data, id_vars=['name'], var_name='type', 
        value_name='val')
```

```python
data1 = pd.DataFrame( {
    'day' : [1,2,3,4,5],
    'wigth' : [95,94,93,94,93],
    'key' : [78,78,78,77,79]
})
data1
```

```python
pd.melt(data1 , id_vars='day')
```

```python
pd.melt ( data1 , id_vars= 'day' , var_name= '01_Col',
         value_name= '02_Col' )
```
